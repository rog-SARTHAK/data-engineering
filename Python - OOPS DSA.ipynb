## Recursion

### Factorial with Recursion

```python
def fac(x):

    if x == 0:
        return 1

    return x * fac(x-1)

print(fac(3))
```

### Fibonacci Number with Recursion

F(0) = 0, F(1) = 1, F(2) = F(0) + F(1)

```python
def fib(n):
    if n == 1:
        value = 1
    elif n == 2:
        value = 1
    elif n > 2:
        value = fib(n-2) + fib(n-1)

    return value

for n in range (1,11):
    print(fib(n))
```

### calculator with class concept

```python
class Calculator:
    def sum(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b != 0:
            return a / b
        else:
            return "Error: Division by zero is not allowed."

# Create an instance of the Calculator class
calc = Calculator()

# Call the methods
result_sum = calc.sum(10, 5)
result_subtract = calc.subtract(10, 5)
result_multiply = calc.multiply(10, 5)
result_divide = calc.divide(10, 5)
result_divide_by_zero = calc.divide(10, 0)

# Print the results
print("Sum: ", result_sum)                # Outputs: Sum:  15
print("Subtract: ", result_subtract)      # Outputs: Subtract:  5
print("Multiply: ", result_multiply)      # Outputs: Multiply:  50
print("Divide: ", result_divide)          # Outputs: Divide:  2.0
print("Divide by zero: ", result_divide_by_zero)  # Outputs: Divide by zero:  Error: Division by zero is not allowed.
```

## `__init__`

`__init__` method is used to initialize a variable. It acts as a constructor. For every object, __init__ gets called automatically.

## Single level Inheritance - OOPS

A <- B

A simple class

```python
class A:

    def feature1(self):
        print ('Feature 1 working')

    def feature2(self):
        print ('Feature 2 working')

a1 = A()

a1.feature1()
a1.feature2()
```

Now let's say we have another class B having some other feature. In class B we also want features of A. This is where we need Inheritance. We will make class B as child class or sub class of A.

```python
class A:

    def feature1(self):
        print ('Feature 1 working')

    def feature2(self):
        print ('Feature 2 working')



class B(A):

    def feature3(self):
        print ('Feature 3 working')

    def feature4(self):
        print ('Feature 4 working')

a1 = A()

b1 = B()

b1.feature1()
b1.feature2()
b1.feature3()
b1.feature4()
```

We see class B has access to all four features. class A will be called Parent class or Super class

## Multi level inheritance - OOPS

A <- B <- C

```python
class A:

    def feature1(self):
        print ('Feature 1 working')

    def feature2(self):
        print ('Feature 2 working')



class B(A):

    def feature3(self):
        print ('Feature 3 working')

    def feature4(self):
        print ('Feature 4 working')


class C(B):

    def feature5(self):
        print ('Feature 5 working')


c1 = C()

c1.feature1()
c1.feature2()
c1.feature3()
c1.feature4()
c1.feature5()
```

## Multiple inheritance - OOPS

A class and B class has no connection. But C class wants to inherit both the features from A and B class

```python
class A:

    def feature1(self):
        print ('Feature 1 working')

    def feature2(self):
        print ('Feature 2 working')



class B():

    def feature3(self):
        print ('Feature 3 working')

    def feature4(self):
        print ('Feature 4 working')


class C(A, B):

    def feature5(self):
        print ('Feature 5 working')


c1 = C()

c1.feature1()
c1.feature2()
c1.feature3()
c1.feature4()
c1.feature5()
```

## Binary Search

List has to be sorted for binary search. It is more efficient than linear search

For a list with index [ 0, 1, 2, 3, 4 ] Binary search has lower index 0, upper index 4. Mid index will be (lower + upper) / 2 = 2 here. Match if search element n == mid value. If its not a match, we have to change lower bound or upper bound. If n is smaller than mid, change upper bound and mid becomes new upper bound. And vice versa. Note: In python // is integer division.

```python
pos = -1 #global variable


def search(list, n):

    l = 0
    u = len(list) - 1

    while l < u :
        mid = (l+u) // 2

        if list[mid] == n:
            globals()['pos'] = mid
            return True

        else:
            if list[mid] < n:
                l = mid
            else:
                u = mid - 1 # - 1 to skip mid value
    return False

list = [5, 8, 10, 20]
n = 8

if search(list, n) :
    print("Found at", pos+1)

else:
    print("Not Found")
```

## Bubble Sort

Say a list [5, 3, 8, 6, 7, 2]

In each iteration,

We take first two values and compare. We swap if former is greater than later.

Say 5 and 3. 5 > 3. So we swap position. New list is [3, 5, 8, 6, 7, 2]

In next iteration, We compare 5, 8. 5 is not > 8. No swap.

After end of first iteration we get [3, 5, 6, 7, 2, 8]

And then starts iteration 2.

We would need 2 loops. Outer loop will take care of iteration. Inner loop will deal with swapping.

range(start, stop, step)

```python
def sort(lst):

    for i in range(0, len(lst)):

        for j in range(i):

            if lst[j] > lst[j+1]:

                temp = lst[j]
                lst[j] = lst[j+1]
                lst[j+1] = temp

lst = [5, 3, 8, 6, 7, 2]
sort(lst)

print(lst)
```

```python
def sort(lst):

    for i in range(len(lst) - 1):
        for j in range(len(lst) - 1):
            if lst[i] > lst[i+1]:
                lst[i], lst[i+1] = lst[i+1], lst[i]


lst = [5, 3, 8, 6, 7, 2]
sort(lst)

print(lst)
```

## Quick Sort

Say there is a list [11, 9, 29, 7, 2, 15, 28]

We try to put pivot in right position that is [7, 9, 2, 11, 29, 15, 28]

This position is placed such that numbers on left of pivot are all smaller than pivot and that on right are all greater.

Now we focus only on left side of pivot

Now, we have an array [7, 9, 2]

7 will be pivot now. We will again try to put pivot in right position.

So now we have [2, 7, 9]

Hence now [2, 7, 9, 11] are sorted already.

Now we look right hand side.

So focus is [29, 15, 28]

29 will be pivot

So we put 29 in right position [15, 28, 29]

Again 15 is pivot.

position of 15 is already right

so we get [2, 7, 9, 11, 15, 28, 29]

This is a divide and conquer problem.

## Singly Linked List - Insertion

image.png

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # Insert at the beginning
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Insert at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    # Insert after a specific node
    def insert_after_node(self, prev_node_data, data):
        if self.head is None:
            print("List is empty")
            return
        current = self.head
        while current is not None:
            if current.data == prev_node_data:
                new_node = Node(data)
                new_node.next = current.next
                current.next = new_node
                return
            current = current.next
        print("Previous node not found")

    # Print the linked list
    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Example usage:
sll = SinglyLinkedList()
sll.insert_at_beginning(10)
sll.insert_at_end(20)
sll.insert_after_node(10, 15)

sll.print_list()

```
