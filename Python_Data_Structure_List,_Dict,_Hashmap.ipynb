# ðŸ” Binary Search

NOTE: List has to be sorted



	1.	Start with two pointers:

	â€¢	lower â†’ at the first index (0)
	â€¢	upper â†’ at the last index (len(List) - 1)

	2.	Find the midpoint: mid = (lower + upper) // 2

	3.	Compare the middle element (List[mid]) with the target (ToSearch):

	â€¢	If equal, the element is found â€” return mid.
	â€¢	If the target is smaller, move the upper bound left: upper = mid - 1
	â€¢	If the target is larger, move the lower bound right:lower = mid + 1

	4.	Repeat the process until lower > upper (which means the element isnâ€™t found).

```python
List = [4, 7, 8, 12, 45, 99]
element = 4
lower = 0
upper = len(List) - 1
position = -1  # default if not found

while lower <= upper:
    mid = (lower + upper) // 2
    if List[mid] == element:
        position = mid
        break
    elif List[mid] < element:
        lower = mid + 1 #target is to the right

    else:
        upper = mid - 1 #target is to the left

print("Position:", position)
```

# Bubble Sort

Screenshot 2025-10-09 at 12.34.45â€¯AM.png

For Bubble sort, we will need the concept of Swapping here. For a given list, we will compare first two elements of the list. If first one is greater than second, we swap the values. We do this step iteratively untill we get sorted list. We will need two loops, one will make sure the biggest element is at the end, and other outer loop will do the same step repeatedly.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):  # repeat n times
        for j in range(0, n - i - 1):  # last i elements already sorted
            if arr[j] > arr[j + 1]:    # compare adjacent elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # swap if needed
    return arr

# Example
nums = [5, 3, 8, 4, 2]
print("Before sorting:", nums)
bubble_sort(nums)
print("After sorting:", nums)
```

ðŸ§  What bubble sort does:

In each pass (outer loop), bubble sort compares adjacent elements and pushes the largest one to the end of the list.

Why "for j in range(0, n - i - 1): " ?

	â€¢	n â†’ total elements
	â€¢	i â†’ how many passes weâ€™ve already done (how many largest elements are already placed at the end)
	â€¢	n - i - 1 â†’ we donâ€™t need to check the last i elements again, because theyâ€™re already sorted.

# Remove duplicates in a list without set() function Leetcode 26

Leetcode 26. Remove Duplicates from Sorted Array (Tests Two-pointer logic and In-place update)

NOTE: Requires list to be sorted

The first value in list must stay in that position. Right pointer will scan through list, left pointer will tell where we will put next unique value. Every time we see a unique value, we increment left pointer by 1.


[0, 0, 1, 1, 1, 2, 2 , 3, 3, 4]

Initialize Left and Right pointer to 2nd position ie., index 1. As First 0 will stay as it is. Right pointer will now check with previous value. It's same. So Right moves next. Left is still second 0. Now Right will compare with Left. This time it's 0 and 1 ie., different. So we put Right pointer value to Left pointer. Next Left pointer shifts to next along with right pointer. Again Right pointer checks the next value.

```python
def remDuplicates(nums):
    if not nums:
        return 0

    left = 1  # pointer for unique elements
    for right in range(1, len(nums)):
        if nums[right] != nums[right - 1]:
            nums[left] = nums[right]
            left += 1
    return left

nums = sorted([int(x) for x in input("Enter numbers: ").split()])
length = remDuplicates(nums)
print("After removing duplicates:", nums[:length])
print("New list length:",length)
```

The goal of your function

You want to remove duplicates in-place â€” meaning:

	â€¢	You donâ€™t create a new list.
	â€¢	You overwrite the original list (nums) to keep only the unique elements at the start.


We use two pointers:

	â€¢	right â†’ scans through the list
	â€¢	left â†’ points to the next position where a unique element should be written

When we find a new unique element, we put it at nums[left] and increment left.

```python
def int_to_rom(x):
    numbers = [1,4,5,9,10,40,50,90,100,400,500,900,1000]
    roman = ['I','IV','V','IX','X','XL','L','XC','C','CD','D','CM''M']
    i = 12 #largest element index of list
    roman_numeral = '' #empty string holding answer
    while x != 0:
        if numbers[i] <= x:
            roman_numeral += roman[i]
            x = x - numbers[i]
        else:
            i = i-1 #move down to previous value
    return roman_numeral

print(int_to_rom(49))
```

Removing duplicate from list. Does preserve order

```python
def remove_duplicates(original_list):
    seen = set()   #set seen to keep track of elements that have already been seen
    seen_add = seen.add     #.add function is used to add elements to the set,
                            #and the list comprehension only includes elements that are not in the set seen
    return [x for x in original_list if not (x in seen or seen_add(x))]



original_list = [4, 6, 8, 4, 9]
print(remove_duplicates(original_list))


#By using the seen_add function within the list comprehension, the order of elements is preserved.
#.add is a method used to add an element to a set
```

 A list comprehension in Python is a compact way to create a new list based on existing lists.

The expression [x for x in original_list if not (x in seen or seen_add(x))] creates a new list with elements from the original_list that are not in the set seen.

For each element x in the original_list, the expression not (x in seen or seen_add(x)) is evaluated. If x is not in the set seen, the seen_add function is called with x as an argument to add it to the set. The not operator inverts the truth value of the expression inside the parenthesis, so if x is not in the set seen, the expression evaluates to True, and the current element x is included in the new list. If x is already in the set seen, the expression evaluates to False, and the current element x is not included in the new list.

The resulting list comprehension returns a list with all unique elements from the original_list, preserving the order of elements. The return statement at the end of the function returns the final list to the caller.

# Linked List

In a linked list, each element (called a node) is stored at an independent memory location. Unlike arrays, which store values in contiguous memory, linked lists store values at random memory addresses. Each node contains two parts:
The actual value.


A pointer (or reference) to the next node in the sequence.


Because of these pointers, the elements are â€œlinkedâ€ together, forming a chain. The first node points to the second, the second points to the third, and so on.
When inserting a new value, we donâ€™t need to shift or copy existing elements like in an array. Instead, we simply:
Create a new node.


Update the pointer of the preceding node to reference this new node.


Make the new node point to the next node in the sequence.


This makes insertion very efficient (O(1) if we already have the pointer to the position), since it only requires modifying a couple of links, not moving large chunks of data.

A doubly linked list improves on a singly linked list by storing not just a link to the next element, but also a link to the previous element. This makes traversal more flexible: we can move forward as well as backward through the list with ease.
On the other hand, arrays have a major advantage over linked lists when it comes to random access. If we know the index of an element in an array, we can access it in constant time (O(1)). For example, retrieving the fifth element is as simple as calling:

array[5]


This operation is immediate because arrays use contiguous memory allocation, so the position of any element can be calculated directly using its index.


### Program for Traversal and printing in Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# Create nodes
head = Node("Orange")
second = Node("Banana")
third = Node("Carrot")

# Link nodes
head.next = second
second.next = third

# Traverse and print
temp = head
while temp:
    print(temp.data)
    temp = temp.next #iterating one by one and print value
```
